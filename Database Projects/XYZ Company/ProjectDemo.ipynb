```python
# IDE demo

import mysql.connector
from tabulate import tabulate
import sys
from IPython.display import clear_output

def clear_screen():
    clear_output(wait=True)

def connect_db():``
    return mysql.connector.connect(
        host="localhost",
        user="xyz_user_demo",
        password="UTDtest",
        database="xyz_company"
    )

def run_view(view_name):
    clear_screen()
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(f"SELECT * FROM {view_name}")
    results = cursor.fetchall()
    print(f"\nResults from: {view_name}", flush=True)
    print(tabulate(results, headers=cursor.column_names, tablefmt="grid"))
    cursor.close()
    conn.close()
    clear_screen()

def display_menu():
    print("\n--- Main Menu ---", flush=True)
    print("1. Views")
    print("2. Stored Procedure")
    print("3. Exit")

def views_menu():
    while True:
        print("\n--- Saved Views ---", flush=True)
        print("a. Average Monthly Salary per Employee")
        print("b. Rounds Passed Per Interviewee Per Job")
        print("c. Items Sold per Product Type")
        print("d. Part Purchase Cost Per Product")
        print("e. Salary Transaction Summary (Derived)")
        print("f. Product Sales Summary (Derived)")
        print("g. Candidate Selection Dates (Derived)")
        print("h. Back to Main Menu")

        print("Select a view: ", end='', flush=True)
        choice = input().lower().strip()

        if choice == 'a':
            run_view("Average_Monthly_Salaries")
        elif choice == 'b':
            run_view("Interviewee_Rounds_Passed")
        elif choice == 'c':
            run_view("Items_Sold_Per_Product_Type")
        elif choice == 'd':
            run_view("Minimum_Parts_Cost_Per_Product")
        elif choice == 'e':
            run_view("Salary_With_Transaction")
        elif choice == 'f':
            run_view("Product_Sales")
        elif choice == 'g':
            run_view("Select_Candidate_Date")
        elif choice == 'h':
            clear_screen()
            break
        else:
            clear_screen()
            print("Invalid choice. Please try again.")
            
def run_stored_procedure(proc_name, params=None):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.callproc(proc_name, params or [])
    for result in cursor.stored_results():
        rows = result.fetchall()
        if rows:
            clear_screen()
            print(f"\nResults from procedure: {proc_name}")
            print(tabulate(rows, headers=result.column_names, tablefmt="grid"))
        else:
            clear_screen()
            print(f"\nNo results returned from {proc_name}.")
    cursor.close()
    conn.close()


def stored_procedure_menu():
    print("\n--- Stored Procedures ---", flush=True)
    print("a. Interviewers for Diana Howard on Job 30018")
    print("b. Marketing Job Postings in July 2024")
    print("c. Employees with No Supervisees")
    print("d. Sites with No Sales in May 2025")
    print("e. Jobs with No Hires After Posting")
    print("f. Sales Reps with Sales Over $200")
    print("g. Departments with No Postings (May–June 2024)")
    print("h. Applicants for Job 30002")
    print("i. Best-Selling Product Type")
    print("j. Most Profitable Product Type")
    print("k. Employees with Multiple Departments")
    print("l. Selected Interviewee Emails (5+ Grades ≥ 60)")
    print("m. Contact Info for Selected Interviewees")
    print("n. Employee with Highest Avg Monthly Salary")
    print("o. Vendor with Lowest Price for Casing 7")
    print("p. Back to Main Menu")

    print("Select a procedure: ", end='', flush=True)
    choice = input().lower().strip()

    proc_map = {
        'a': "sp_interviewers_for_candidate_job",
        'b': "sp_marketing_jobs_july2024",
        'c': "sp_employees_no_supervisees",
        'd': "sp_sites_no_sales_may2025",
        'e': "sp_jobs_no_hires_after_posting",
        'f': "sp_sales_over_200_by_sales_rep",
        'g': "sp_departments_no_postings_may_jun_2024",
        'h': "sp_applicants_for_job_30002",
        'i': "sp_best_selling_product_type",
        'j': "sp_most_profitable_product_type",
        'k': "sp_employees_with_multiple_departments",
        'l': "sp_selected_interviewees_emails",
        'm': "sp_selected_interviewee_contact_info",
        'n': "sp_employee_with_highest_avg_salary",
        'o': "sp_lowest_price_vendor_for_casing7",
    }

    if choice in proc_map:
        clear_screen()
        run_stored_procedure(proc_map[choice])
        input("\nPress Enter to return to the main menu...")
    elif choice == 'p':
        clear_screen()
        return
    else:
        clear_screen()
        print("Invalid choice.")


########### Main ###########            
def main():
    print('Welcome to the MySQL/Python demonstration for the XYZ Company Database.', flush=True)

    while True:
        display_menu()
        print("Select an option: ", end='', flush=True)
        choice = input().lower().strip()

        if choice == '1':
            views_menu()
            clear_screen()
        elif choice == '2':
            stored_procedure_menu()
            clear_screen()
        elif choice == '3':
            print("Exiting the interface. Goodbye!")
            clear_screen()
            break
        else:
            clear_screen()
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()

```
